The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python

# Add marker_cluster to current site_map
# Create a map centered at the first launch site's coordinates
site_map_2 = folium.Map(location=[spacex_df['Lat'].iloc[0], spacex_df['Long'].iloc[0]], zoom_start=10)

# Create a marker cluster
marker_cluster = MarkerCluster().add_to(site_map_2)

# Function to determine marker color based on the class value
def get_marker_color(class_value):
    return 'green' if class_value == 1 else 'red'

# Iterate over each row in the dataframe
for index, record in spacex_df.iterrows():
    # Get the coordinates, class value, and launch site name
    coordinate = [record['Lat'], record['Long']]
    class_value = record['class']
    launch_site = record['Launch Site']
    
    # Determine the marker color based on the class value
    marker_color = get_marker_color(class_value)
    
    # Create a Marker object with customized icon and label
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color='white', icon_color=marker_color),
        tooltip=launch_site

    )
    
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)
    
    marker_site = folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % launch_site,
        )
    )
    site_map_2.add_child(marker_site)
    
    # Create a circle with a popup for the launch site
    circle = folium.Circle(coordinate, radius=50, color='#d35400', fill=False).add_child(folium.Popup(name))
    site_map_2.add_child(circle)

# Display the map
site_map_2
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example

# find coordinate of the closet coastline
# e.g.,: site CCAFS SLC-40(28.562302,-80.577356), costalline point (28.562302, -80.576820)
# Define the coordinates of the two points
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# Function to calculate distance between two points using haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # approximate radius of Earth in km

    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Define the coordinates of the launch site and coastline points
site_coordinates = [28.562302, -80.577356]
coastline_points = [
    [28.56232, -80.56766],
    [28.57155, -80.58506],
    [28.56213, -80.57065],
    [28.39918, -80.60634]
]

site_map = folium.Map(location=site_coordinates, zoom_start=13)

# Add a circle representing the launch site
circle = folium.Circle(site_coordinates, radius=50, color='#d35400', fill=True)
site_map.add_child(circle)

# Add a marker for the launch site
site_marker = folium.Marker(
    location=site_coordinates,
    icon=folium.DivIcon(
        icon_size=(150,36),
        icon_anchor=(75,18),
        html='<div style="font-size: 12px; color:#d35400;"><b>CCAFS LC-40</b></div>'
    )
)
site_map.add_child(site_marker)

# Connect the launch site to coastline points with lines and display distance
for point in coastline_points:
    line_coordinates = [site_coordinates, point]
    folium.PolyLine(locations=line_coordinates, color='red', weight=2.5, dash_array='5, 5').add_to(site_map)
    distance = calculate_distance(site_coordinates[0], site_coordinates[1], point[0], point[1])
    distance_marker = folium.Marker(
        location=point,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance)
        )
    )
    site_map.add_child(distance_marker)

# Display the map
site_map
```
